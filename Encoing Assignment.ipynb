# Label Encoder

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df1 = df.copy()
df2 = df.copy()
df3 = df.copy()
```

```python
from pandas.core.dtypes.common import is_numeric_dtype
```

```python
df.columns
```

```python
df1.head()
```

```python
#loop

for col in df1.columns:
    if is_numeric_dtype(df1[col]):
        continue
    else:
        df1[col] = le.fit_transform(df1[col])
```

# Ordinal

```python
df5 = df.copy()
```

```python
df5.head()
```

```python
df5.region.unique()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
area = df5.region.unique()
order = OrdinalEncoder(categories=[area])
encoded = order.fit_transform(df5[['region']])
```

```python
df5['region'] = encoded 
```

```python
df5.head()
```

```python
for col in df5.columns:
    if is_numeric_dtype(df5[col]):
        continue
    else:
        df5[col] = le.fit_transform(df5[col])
```
